## Vocabulario (palabras, bigrams y trigrams) por categoría

```python
from nltk import ngrams
from nltk.probability import FreqDist
from collections import Counter
```

```python
def get_vocabulary(df, col_name):
    
    words_ = []
    bigrams_ = []
    trigrams_ = []
    
    for review in df[col_name]:
        for word in review.split():
            words_.append(word)
    
    bigrams_ = list(ngrams(words_, 2))
    trigrams_ = list(ngrams(words_, 3))
    
    vocab = {
        'words': words_,
        'bigrams': bigrams_,
        'trigrams': trigrams_
    }
    
    return vocab
```

## Riqueza léxica por categoría

La riqueza léxica de un documento (o corpus) es una medida de la varianza en el léxico utilizado. Dados dos documentos de igual extensión (número total de términos) aquel con mayor número de palabras (vocabulario con mayor cardinalidad) tendrá mayor riqueza léxica.

Existen diversas maneras de calcular la riqueza léxica. La más sencilla es dividiendo el número total de términos únicos entre el total de términos. Otras maneras de calcularla tienen en cuenta la categoría gramatical de los términos, si son informativos o no, etc.

A continuación aparece calculada la riqueza léxica para la reviews de cada categoría.

# Preprocesamos el texto

Con el objetivo de normalizar el texto de las reviews, así como de trabajar con un vocabulario menor que el original que contenga menos palabras "poco informativas", la etapa de procesado de texto propuesta consiste en:

- Conversión de todo el texto a minúsculas
- Eliminación de números (dígitos)
- Eliminación de símbolos
- Eliminación de signos de puntuación
- Lemmatización sin desambiguación (búsqueda directa en diccionario de formas flexionadas y lemmas)

```python
import string
import unicodedata
from stop_words import get_stop_words
```

```python
def text_normalizer(text, lemmas_dict=None, sw_list=None):

    processed_text = []
    
    if sw_list is None:
        sw_list = []
    
    # Convierte el texto a minúsuculas
    text = text.lower()
    
    # Elimina números (dígitos)
    text = ''.join([c for c in text if not c.isdigit()])
    
    # Eliminar caracteres "extraños"
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    
    # Tabla para eliminar signos de puntuación
    table = str.maketrans('', '', string.punctuation)
    
    # Segmentar texto en frases
    sentences = text.split('.')
    
    # Para cada frase
    for sentence in sentences:
        words = sentence.split(' ')
        # Para cada palabra
        for word in words:
            if word not in sw_list:
                if word not in string.punctuation:
                    word = word.translate(table)
                    if lemmas_dict is not None:
                        word = lemmas_dict.get(word, word)
                processed_text.append(word)
    
    return ' '.join(processed_text)
```

```python
def get_lemmas_dict(data_path, lemmas_dict_file):
    lemmas_dict = {}
    with open(os.path.join(data_path, lemmas_dict_file), 'r') as f:
        for line in f:
            (key, val) = line.split()
            lemmas_dict[str(val)] = key
    return lemmas_dict
```

```python
lemmas_path = '../data'
english_lemmas = 'lemmatization-en.txt'  # Fuente: https://github.com/michmech/lemmatization-lists

english_lemmas_dict = get_lemmas_dict(lemmas_path, english_lemmas)
```

```python
sw_list = get_stop_words('english')
```

```python
df_automotive['processed_text'] = df_automotive['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
df_clothing['processed_text'] = df_clothing['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
df_office['processed_text'] = df_office['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
df_sports['processed_text'] = df_sports['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
df_videogames['processed_text'] = df_videogames['reviewText'].apply(lambda text: text_normalizer(text=text, lemmas_dict=english_lemmas_dict, sw_list=sw_list))
```

# Entrenamos un LDA para el modelado de topics

## Diccionario id-término y matriz documento-palabra

```python
import gensim
from gensim.corpora import Dictionary
from gensim.models import LdaModel, CoherenceModel

import pyLDAvis
import pyLDAvis.gensim
```

```python
all_reviews_text = \
    list(df_automotive['reviewText'].values) + \
    list(df_clothing['reviewText'].values) + \
    list(df_office['reviewText'].values) + \
    list(df_sports['reviewText'].values) + \
    list(df_videogames['reviewText'].values)

all_reviews_text = [review.split() for review in all_reviews_text]
```

```python
all_reviews_processed_text = \
    list(df_automotive['processed_text'].values) + \
    list(df_clothing['processed_text'].values) + \
    list(df_office['processed_text'].values) + \
    list(df_sports['processed_text'].values) + \
    list(df_videogames['processed_text'].values)

all_reviews_processed_text = [review.split() for review in all_reviews_processed_text]
```

```python
raw_dictionary = Dictionary(all_reviews_text)
processed_dictionary = Dictionary(all_reviews_processed_text)
```

```python
raw_corpus = [raw_dictionary.doc2bow(doc) for doc in all_reviews_text]
processed_corpus = [processed_dictionary.doc2bow(doc) for doc in all_reviews_processed_text]
```

## Entrenamiento

Elegimos un número de topics igual a cinco ya que es número de categorías que estamos usando como corpus. A priori debería ser el número de topics óptimo.

```python
num_topics = 5
```

```python
lda_model_raw = LdaModel(
    corpus=raw_corpus,
    id2word=raw_dictionary,
    num_topics=num_topics,
    iterations=5,
    passes=10,
    alpha='auto'
)
```

```python
word_dict_raw = {}

for i in range(num_topics):
    words = lda_model_raw.show_topic(i, topn = 20)
    word_dict_raw['Topic #' + '{:02d}'.format(i+1)] = [i[0] for i in words]
pd.DataFrame(word_dict_raw)
```

Con el texto sin procesar se antoja imposible la identificación de ningún topic.

```python
lda_model_processed = LdaModel(
    corpus=processed_corpus,
    id2word=processed_dictionary,
    num_topics=num_topics,
    iterations=5,
    passes=10,
    alpha='auto'
)
```

```python
word_dict_processed = {}

for i in range(num_topics):
    words = lda_model_processed.show_topic(i, topn = 15)
    word_dict_processed['Topic #' + '{:02d}'.format(i+1)] = [i[0] for i in words]
pd.DataFrame(word_dict_processed)
```

# Buscamos un número de topics - teóricamente - mejor

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        
        # Build LDA model
        model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=dictionary,
                                                num_topics=num_topics)
        
        # Create a list of LDA models
        model_list.append(model)
        
        # Compute the Coherence for each model
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
start_ = 1
end_ = 20
step_ = 1
```

```python
model_list, coherence_values = compute_coherence_values(
    dictionary=processed_dictionary,
    corpus=processed_corpus,
    texts=all_reviews_processed_text,
    start=start_,
    limit=end_,
    step=step_
)
```

```python
x = range(start_, end_, step_)
plt.plot(x, coherence_values)
plt.xlabel('Num Topics')
plt.ylabel('Coherence score')
plt.legend(('coherence_values'), loc='best')
plt.show()
```

```python
optimal_model = model_list[3]

word_dict = {};
for i in range(4):
    words = optimal_model.show_topic(i, topn = 10)
    word_dict['Topic #' + '{:02d}'.format(i+1)] = [i[0] for i in words]
pd.DataFrame(word_dict)
```

```python
# Perplexity
print('\nPerplexity ("optimal" model): ', optimal_model.log_perplexity(processed_corpus))

# Coherence Score
coherence_optimal_model_processed = CoherenceModel(model=optimal_model, texts=all_reviews_processed_text, dictionary=processed_dictionary, coherence='c_v')
coherence_optimal_processed = coherence_optimal_model_processed.get_coherence()
print('\nCoherence Score ("optimal" model): ', coherence_optimal_processed)
```

Puede verse como en los entrenamientos para distintos números de topics no es cinco el valor con mejores métricas. Es, en cambio, cuatro, el número que parece optimizar los resultados.

Al no ser un modelo determinista diferentes inicializaciones pueden arrojar diferentes resultados.

Atendiendo al valor del perplexity y del coherence para cuatro como número de topics, así como a la variación de estos valores para distintos valores en el número de topics, no podemos determinar que exista un número claramente mejor que otro. Será nuestro conocimiento del "negocio" el que nos permita elegir un valor u otro. Como conocemos a priori las categorías, y estamos asumiendo que un topic representará una categoría (en la exploración pudimos ver como el vocabulario por categoría difería del resto cuando se procesaba el texto), determinaremos que cinco será el valor óptimo para entrenar el LDA.

Pese a esto, evaluamos los resultados de la visualización para el modelado de topics con cuatro topics.
